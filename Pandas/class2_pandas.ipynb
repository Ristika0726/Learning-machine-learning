##Data Cleaning with pandas ðŸ¼

```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/College/5th sem/pandas_dataset.csv")
data.head(5)
```

```python
data["Date"]=data["Date"].str.replace("'","")
print(data.head(5))
```

```python
data["Date"]=data["Date"].replace("'","",regex=True)
data.head(5)
```

```python
data
```

```python
#return a dataset with no empty cells
#data.dropna() will remove empty cells but will not change the dataframes
#if we want data frames to be changed, use inplace=true
new_data=data.dropna()
new_data
```

```python
#another way of handling the null values id by replacing the null values with some other values
data_new2=data.fillna(130)
data_new2
```

```python
#Fill specific column's null values
data
```

```python
data_new3=data
data_new3["Calories"].fillna(150,inplace=True)
data_new3
```

```python
#fill specific null values with mean or median or more
import pandas as pd
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/College/5th sem/pandas_dataset.csv")
x=dataset["Calories"].mean()
dataset["Calories"].fillna(x, inplace=True)
dataset
```
